从开始说要离职到现在,我居然连简历都没有投几封,看来整个人变得越来越懒惰了.不管怎样,还是打起精神,不管有多么的疲倦,还是要继续找工作的,懒惰,只会让人越来越懒,一事无成.

上次的自我总结,说道了第一家创业公司时工作的种种问题,接下来继续说后面的事情吧.

离开了创业公司,那时我刚刚毕业,经过短暂的修整之后,开始了找工作之旅.此次找工作,我觉得自己刚刚毕业,毕竟经验不是很丰富,许多方面需要提升,因此将自己的定位放得比较的低,但是还是太嫩了,我被忽悠进了培训机构,浪费了整整半年的时间,之后自己在培训机构没有学到任何的东西,靠自己以前的积累,还是找到了一份工作.这份工作干得相当的累,基本上去年我没有怎么双休,一直加班,直到去年年底才辞职.在这份工作的期间,我发明我的能力其实和大多数编写java的开发人员差不多,有时候甚至比他们强一些,但是呢,觉得工作没有什么意思,我一直认为工作应该好好的干,自己应该有与工作相匹配的能力.然而现实却是,公司不需要,公司需要的是廉价的劳动力,所以,只要能把事情干成,找几个廉价劳动力又如何,说不定可以培养成功呢.另外,这个公司一直死命的加班,在我看来,如果找几个能力强的,也许要不了这么多人,这个项目也可以完成,然而却不是.这一点,一直与我的价值观相悖,我也喜欢这家公司,最后还是跳槽走人了.

在这家公司待了一段时间,我发现自己还是有一些的缺点,比如,有能力,但是事情做不好.做事有时候太拖沓,缺乏沟通与交流,做出的东西与实际的要求大相庭径,缺乏主动性,等等问题.这个也许是我一直有能力,却无法把事情做好的原因吧.
