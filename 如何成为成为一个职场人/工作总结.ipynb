


# it开发两年工作总结

>最近一个朋友给我说,我们在换工作的时候要懂得复盘,明白在一家公司里面我们能够得到什么,自己每一次的跳槽是在一个公司对个人的成长已经没有价值了,需要更高的平台的时候,才需要换一份新的工作.我觉得他说的话非常有道理,加上最近想换工作了,决定总结一下自己的过去,看一看自己在最近一年中的收获和成长.
从去年毕业到现在以来,总的来说已经找了4次工作了,总的来说,个人认为这几次工作都是比较失败的.或者说从实习到现在,我换的每一份工作,个人觉得价值和意义都不是太有意义.但是,在这个过程中还是学到了许多的经验和教训,也逐渐明白自己应该如何发展,个人的未来应该如何规划了.

>第一份工作是在朋友的公司做实习,其实朋友待我还是挺不错的,但是从个人的发展来说,其实我当时应该仔细的考虑一下自己的未来,再做决定的.当时,我的个人来说是比较迷茫的,因为我当时个人状态不是太好,而且整个人也有一点自闭,整个人就是一团乱,在这种状态下,我在朋友的公司工作了大概一年的样子.在这家公司,我们主要是做一款自己的笔记类app的开发,我主要负责后端的开发工作.这一段时间的开发,由只有一个工作了两年左右的.net开发和一个测试转开发的ios,一个半吊子前端和两个毕业的实习生,一个是我做后端,另一个做android,招聘了一个培训班出来的ui,产品是我的朋友,和我一样是刚毕业的.我和朋友大学的时候就在做开发,一起合作,从整体上来说,我们都是新手,没有什么经验的一个草草组成的团队.我们一起开发了大概6个月左右,完成了产品的一个出版,相比起我们预计的3个月完成,多花了一倍的时间.然后产品上传应用宝,开始推广,目前用户大概有几万人吧,没有太过关注这个.

>总结起来,大概有以下几点收获和经验吧

## 自我迷茫,不知道方向和未来

其实我在朋友公司,我一点也不想待,而且也认为公司的项目没有什么发展,个人也不认为创业能够成功.但是出于一些个人的原因,我一直想戒掉一些不良习惯,所以一直把我待在朋友公司作为一种类似于惩罚的措施,但是到了离开公司,我也没有改变自己的习惯,反而对自身的影响越来越严重,整个人对生活中的其他事情也越来越消极了.

## 过于看重工资

当时朋友给我开的工资是普通实习生工资的两倍,当时只顾眼前利益,觉得工资高,而没有考虑自己的将来,所以也就半推半就的留下来了,目光有一些短浅.

## 盲目自信,过于高看自己

其实项目进展缓慢,从我在大学的时候就发现了这个问题了,但是当时比较在上大学,以为是上学的原因,没有足够的时间,所以项目进展缓慢,但是其实从另一个角度来说,个人认为是因为我们经验缺乏,没有做合理的规划.不过,我估计朋友把这个算进去了吧,因为从一开始是的时候,他大概就是打算让我们边研究,边做东西吧.不过,这个不是我喜欢的方式,我认为做事就应该求稳,几个新手经验不足,慢慢摸索,不喜欢.而我实际做的时候,的确把自己当成一个经验丰富的开发者来对待的,然后快速的完成了后端的开发工作,基本上算是按时完成,不过android端的开发和我预料的差不多,进展缓慢,一拖再拖.而ios端,情况也没有好多哪里去.不过这个,貌似又不能算我高看自己,只能说自己无法把自己的想法,和里表达出来,成为帮助团队发展的动力.

## 浪费了自己的能力

当时我对android开发比较了解,而且对后端开发也比较熟悉,但是没有站出来,在团队中发挥自己的特长,而是得过且过,慢慢的消磨时光,简直是浪费.
