```python
from pyspark import SparkConf
from pyspark.streaming import StreamingContext
import time 
ssc = StreamingContext(sc, 1)
lines=ssc.textFileStream("file:///E:/notebook/genlog/stream")
```

```python
# 简单的wordcount测试
words=lines.flatMap(lambda line:line.split(' '))
pairs=words.map(lambda word:(word,1))
wordCounts = pairs.reduceByKey(lambda x, y: x + y)
```

```python
def convert(rdd):
    return rdd.map(lambda word:(word[1],word[0])).sortByKey(False).map(lambda ip_pv:(ip_pv[1],ip_pv[0]))
```

```python
lines.map(lambda line:(line.split(' ')[0],1))\
.reduceByKey(lambda a,b:a+b).\
transform(convert).pprint()
```
